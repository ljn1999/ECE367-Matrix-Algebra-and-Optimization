```python
A = np.asarray([[1, 1], [0, 1]])
b = np.asarray([1/2, 1])

y = np.asarray([1, 0])
_A_ = np.zeros((2, 10))
for i in range(5):
  product = np.identity(2)
  for j in range(0, i):
    product = product @ A
  product = product @ b
  _A_[:,5-1-i] = product
```

```python
C = np.concatenate((A_, _A_[0:1, :]))
y_new = np.asarray([1, 0, 0])
p_star = C.T @ np.linalg.inv(C @ C.T) @ y_new
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
time = [str(i) for i in range(1, 11)]
ax.bar(time,p_star)
plt.title("Plot of f(n) in (c)")
plt.xlabel("time")
plt.ylabel("force")
plt.show()
```

```python
# velocity is linear
v = [0]
for slot in range(1, 11):
  v.append(v[-1] + p_star[slot-1])

plt.plot(range(0, 11), v)
plt.title("Plot of x dot (t) in (c)")
plt.xlabel("time")
plt.ylabel("velocity")
plt.show()
```
